# explanatory data analysis

```python
x = [1, 2, 3]
victim_1 = ['cyclist', 'passenger', 'pedestrian']
plt.pie(x, labels=victim_1, autopct='%.1f%%')
plt.show()
```

```python
print(df.columns)
freq_table = pd.crosstab(df['victim_1'], df['accident'], margins=True, margins_name='Frequency')
print(freq_table)
```

```python
# Calculate percentages for each victim type per accident
freq_table_pct = freq_table.div(freq_table['Frequency'], axis=0) * 100
```

```python
# Print the table with percentages
print(freq_table_pct)
```

```python
# Calculate percentages for each victim type per accident type
head_on_pct = [freq_table.loc['cyclist', 'Head_oc'] / freq_table.loc['Frequency', 'Head_oc'] * 100,
              freq_table.loc['passenger', 'Head_oc'] / freq_table.loc['Frequency', 'Head_oc'] * 100,
              freq_table.loc['pedestrian', 'Head_oc'] / freq_table.loc['Frequency', 'Head_oc'] * 100]

hit_run_pct = [freq_table.loc['cyclist', 'hit_Run'] / freq_table.loc['Frequency', 'hit_Run'] * 100,
               freq_table.loc['passenger', 'hit_Run'] / freq_table.loc['Frequency', 'hit_Run'] * 100,
               freq_table.loc['pedestrian', 'hit_Run'] / freq_table.loc['Frequency', 'hit_Run'] * 100]

# Create pie charts
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

ax1.pie(head_on_pct, labels=freq_table.index[:-1], autopct='%1.1f%%')
ax1.set_title('Head-on Collision')

ax2.pie(hit_run_pct, labels=freq_table.index[:-1], autopct='%1.1f%%')
ax2.set_title('Hit and Run')

plt.show()
```

```python
contingency_table = pd.crosstab(df['age'], df['accident'])
print(contingency_table)
```

```python
import scipy.stats as stats
chi2, p, _, _ =stats.chi2_contingency(contingency_table)
print(f"chi-square statistic: {chi2}")
print(f"p-value: {p}")
```
